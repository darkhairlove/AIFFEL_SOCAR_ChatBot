```python
import pandas as pd

file = pd.read_csv('socar_data.txt', sep='\t')
file
```

```python
arr = []
for i in file['AA']:
    if i != '=':
        arr.append(i.replace('\xa0',''))
```

```python
arr
```

```python
qn = []
an1 = []
an = ''
for i in arr:
    if '?' in i:
        qn.append(i)
        an1.append(an)
        an = ''
    else:
        an += i
print(len(qn))
print(len(an1))
```

```python
an1.append(file['AA'][1659]+file['AA'][1660])
del an1[0]
print(len(an1))
```

```python
df2 = pd.DataFrame(columns=['Q', 'A'])
for i, v in zip(qn,an1):
    df2 = df2.append(pd.DataFrame([[i,v]], columns=['Q','A']), ignore_index=True)
df2
```

```python
df = pd.DataFrame(columns=['Q', 'A'])
for i in arr:
    if '?' in i:
        df = df.append(pd.DataFrame([[i]], columns=['Q']), ignore_index=True)
    else:
        df = df.append(pd.DataFrame([[i]], columns=['A']), ignore_index=True)
df
```

```python
Q_=df['Q'].ffill()
Q_
```

```python
A_=df.drop(['Q'],axis=1)
A_
```

```python
df_1= pd.concat([Q_, A_], axis=1)
df_1
```

```python
df=df_1.dropna()
df
```
